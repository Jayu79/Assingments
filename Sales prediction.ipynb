```python
from sklearn import datasets
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import VotingClassifier
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import precision_score, recall_score, f1_score, confusion_matrix
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("sales_train_v2.csv")
```

```python
d=df[:10000]
```

```python
d=d.dropna()
```

```python
X = d.drop(labels = ["date","item_cnt_day"],axis=1)
```

```python
X
```

```python
Y = d["item_cnt_day"].values
```

```python
Y
```

```python
x_train, x_test, y_train, y_test = train_test_split(X,Y, test_size=0.2, random_state=12)
```

```python
#averaging
voting = VotingClassifier(estimators=[
    ('Logreg', LogisticRegression()),
    ('KNC', KNeighborsClassifier()),
    ('DTC', DecisionTreeClassifier())
], voting='soft')
```

```python
voting.fit(x_train, y_train)
```

```python
predictions = voting.predict(x_test)
```

```python
predictions
```

```python
voting.score(x_test, y_test)
```

```python
#voting
voting1 = VotingClassifier(estimators=[
    ('Logreg', LogisticRegression()),
    ('KNC', KNeighborsClassifier()),
    ('DTC', DecisionTreeClassifier())
])
```

```python
voting1.fit(x_train, y_train)
```

```python
predictions1 = voting1.predict(x_test)
```

```python
predictions1
```

```python
confusion_matrix(y_test, predictions1)
```

```python
voting1.score(x_test, y_test)
```
